```python
#Завдання 2. Аналіз повноти відповідей респондентів

import pandas as pd

# Завантаження файлів
df = pd.read_csv('results.csv', low_memory=False)
schema = pd.read_csv('schema.csv')

# Отримання списку питань з поля qname
questions = set(schema['qname'])

# Отримання колонок датафрейму
df_columns = set(df.columns)

# Інтерсекція множин:
# залишаємо лише ті питання, які є і в schema, і в results
valid_questions = list(questions.intersection(df_columns))

# Перевірка респондентів без пропущених значень
complete_respondents = df[valid_questions].notna().all(axis=1).sum()

# Фінальний результат
complete_respondents
```

```python
#Завдання 3. Статистичний аналіз досвіду респондентів

import pandas as pd

# Завантаження файлів
df = pd.read_csv('results.csv', low_memory=False)

#Аналіз колонки WorkExp
work_exp = df['WorkExp']

#Обчислення мір центральної тенденції
mean_value = work_exp.mean()
median_value = work_exp.median()
mode_value = work_exp.mode()[0]

#Формування результату у зручному форматі
statistics = pd.Series({
    'Mean': mean_value,
    'Median': median_value,
    'Mode': mode_value
})

#Виведення результату
statistics
```

```python
#Завдання 5. Визначення популярності Python

import pandas as pd

# Завантаження датасету
df = pd.read_csv('results.csv', low_memory=False)

# Перевірка, чи містить колонка LanguageHaveWorkedWith значення "Python"
python_developers = df['LanguageHaveWorkedWith'].str.contains(
    'Python',
    case=False,
    na=False
)

# Обчислення відсотка респондентів, які програмують на Python
python_percentage = (python_developers.sum() / len(df)) * 100

# Виведення результату у відсотковому форматі
round(python_percentage, 2)
```

```python
#Завдання 6. Аналіз шляхів навчання програмуванню

import pandas as pd

# Завантаження датасету
df = pd.read_csv('results.csv', low_memory=False)

# Пошук респондентів, які навчалися через онлайн курси
online_courses = df['LearnCode'].str.contains(
    'Online Courses',
    case=False,
    na=False
)

# Підрахунок кількості респондентів
online_courses_count = online_courses.sum()

# Результат
online_courses_count
```

```python
#Завдання 7. Географічний аналіз компенсації Python-розробників

import pandas as pd

# Завантаження датасету
df = pd.read_csv('results.csv', low_memory=False)

# Фільтрація Python-розробників
python_developers = df[
    df['LanguageHaveWorkedWith'].str.contains(
        'Python',
        case=False,
        na=False
    )
]

# Видалення рядків з пропущеними значеннями компенсації або країни
python_developers = python_developers.dropna(
    subset=['Country', 'ConvertedCompYearly']
)

# Групування по країнах та обчислення статистики
compensation_by_country = python_developers.groupby('Country')[
    'ConvertedCompYearly'
].agg(
    Mean_Compensation='mean',
    Median_Compensation='median'
).reset_index()

# Виведення результату
compensation_by_country
```

```python
#Завдання 8. Аналіз освіти найбільш оплачуваних спеціалістів

import pandas as pd

# Завантаження датасету
df = pd.read_csv('results.csv', low_memory=False)

# Видалення рядків з пропущеними значеннями компенсації або освіти
filtered_df = df.dropna(subset=['ConvertedCompYearly', 'EdLevel'])

# Сортування за компенсацією у спадному порядку
top_5_paid = filtered_df.sort_values(
    by='ConvertedCompYearly',
    ascending=False
).head(5)

# Виведення рівнів освіти топ-5 респондентів
top_5_education = top_5_paid[['ConvertedCompYearly', 'EdLevel']]

# Результат
top_5_education
```

```python
#Завдання 9. Аналіз популярності Python по віковим категоріям

import pandas as pd

# Завантаження датасету
df = pd.read_csv('results.csv', low_memory=False)

# Видалення пропущених значень у віковій категорії
age_df = df.dropna(subset=['Age'])

# Перевірка, чи використовує респондент Python
age_df['UsesPython'] = age_df['LanguageHaveWorkedWith'].str.contains(
    'Python',
    case=False,
    na=False
)

# Групування по вікових категоріях та обчислення відсотка
python_by_age = age_df.groupby('Age')['UsesPython'].mean() * 100

# Перетворення у DataFrame для зручного вигляду
python_by_age = python_by_age.reset_index(name='PythonPercentage')

# Округлення результату
python_by_age['PythonPercentage'] = python_by_age[
    'PythonPercentage'
].round(2)

# Результат
python_by_age
```
